```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, SubsetRandomSampler
from torchvision import datasets, transforms
import random
import numpy as np
import matplotlib.pyplot as plt
import byzfl
import torch.nn.functional as F
import copy

# ------------------------------------------------------------------------------
# Fonctions utilitaires
def merge_intervals(intervals):
    if not intervals:
        return []
    intervals.sort()
    merged = [intervals[0]]
    for current in intervals[1:]:
        last = merged[-1]
        if current[0] <= last[1]:
            merged[-1] = (last[0], max(last[1], current[1]))
        else:
            merged.append(current)
    return merged

def generate_gaussian_availability(total_time_steps, worker_id, num_workers,
                                   std_dev_ratio=0.08, min_processing_time=3,
                                   smoothing_window=10):
    peak_means = [0.2, 0.5, 0.8]
    std_dev = int(std_dev_ratio * total_time_steps)
    intervals = []
    for mean_ratio in peak_means:
        mean = int(mean_ratio * total_time_steps)
        start = int(np.random.normal(loc=mean, scale=std_dev))
        start = max(0, min(start, total_time_steps - min_processing_time*3))
        duration = random.randint(min_processing_time*3,
                                  min(min_processing_time*5, total_time_steps - start))
        intervals.append((start, start + duration))
    merged = merge_intervals(intervals)
    smoothed = []
    for s, e in merged:
        smoothed.append((max(0, s - smoothing_window), min(total_time_steps, e + smoothing_window)))
    return merge_intervals(smoothed)

def generate_gaussian_byzantine_availability(total_time_steps, worker_id, num_workers,
                                             std_dev_ratio=0.05, min_processing_time=3,
                                             smoothing_window=10):
    off_peak_means = [0.1, 0.9]
    std_dev = int(std_dev_ratio * total_time_steps)
    intervals = []
    for mean_ratio in off_peak_means:
        mean = int(mean_ratio * total_time_steps)
        start = int(np.random.normal(loc=mean, scale=std_dev))
        start = max(0, min(start, total_time_steps - min_processing_time*4))
        duration = random.randint(min_processing_time*4,
                                  min(min_processing_time*8, total_time_steps - start))
        intervals.append((start, start + duration))
    merged = merge_intervals(intervals)
    smoothed = []
    for s, e in merged:
        smoothed.append((max(0, s - smoothing_window), min(total_time_steps, e + smoothing_window)))
    return merge_intervals(smoothed)

# ------------------------------------------------------------------------------
# Modèle LeNetLike
class LeNetLike(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool  = nn.MaxPool2d(2, 2)
        self.fc1   = nn.Linear(64*7*7, 128)
        self.dropout = nn.Dropout(0.25)
        self.fc2   = nn.Linear(128, 10)
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64*7*7)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        return self.fc2(x)

# ------------------------------------------------------------------------------
# BatchManager
class BatchManager:
    def __init__(self, dataset, batch_size):
        self.dataset    = dataset
        self.batch_size = batch_size
        self.indices    = list(range(len(dataset)))
        random.shuffle(self.indices)
        self.sampler    = SubsetRandomSampler(self.indices)
        self.loader     = iter(DataLoader(dataset, batch_size=batch_size, sampler=self.sampler))
    def get_next_batch(self):
        try:
            return next(self.loader)
        except StopIteration:
            random.shuffle(self.indices)
            self.sampler = SubsetRandomSampler(self.indices)
            self.loader  = iter(DataLoader(self.dataset, batch_size=self.batch_size, sampler=self.sampler))
            return next(self.loader)

# ------------------------------------------------------------------------------
# Worker / ByzantineWorker
class Worker:
    def __init__(self, model, device, worker_id, availability, processing_time):
        self.is_byzantine   = False
        self.device         = device
        self.worker_id      = worker_id
        self.availability   = availability
        self.proc_time      = processing_time
        self.model          = model.to(device)
        self.criterion      = nn.CrossEntropyLoss()
        self.finish_time    = None
        self.current_info   = None
    def is_available(self, t):
        for s,e in self.availability:
            if s <= t < e and (self.finish_time is None or t >= self.finish_time):
                return (t + self.proc_time) <= e
        return False
    def start(self, t, info, data, targets):
        if self.current_info != info:
            self.current_info = info
            self.finish_time  = t + self.proc_time
            return self.compute(data, targets)
        return None
    def has_finished(self, t):
        return self.finish_time == t
    def compute(self, data, targets):
        data, targets = data.to(self.device), targets.to(self.device)
        self.model.zero_grad()
        out = self.model(data)
        loss = self.criterion(out, targets)
        loss.backward()
        grads = {n: p.grad.detach().clone() for n,p in self.model.named_parameters()}
        return grads, self.current_info
    def update_model(self, sd):
        self.model.load_state_dict(sd)

class ByzantineWorker(Worker):
    def __init__(self, model, device, worker_id, availability, processing_time):
        super().__init__(model, device, worker_id, availability, processing_time)
        self.is_byzantine = True
    def compute(self, data, targets):
        corrupted = {}
        for n,p in self.model.named_parameters():
            corrupted[n] = torch.full_like(p, 1e6)
        return corrupted, self.current_info

# ------------------------------------------------------------------------------
# Server
class Server:
    def __init__(self, dataset, batch_size=64, mini_batch_size=16, lr=0.005,
                 aggregation_strategy='trimmed_mean', preagg=None, verbose=False):
        self.device               = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model                = LeNetLike().to(self.device)
        self.optimizer            = optim.Adam(self.model.parameters(), lr=lr)
        self.batch_mgr            = BatchManager(dataset, batch_size)
        self.mini_batch_size      = mini_batch_size
        self.aggregation_strategy = aggregation_strategy
        self.preagg               = preagg
        self.verbose              = verbose
        self.current_batch_id     = 0
        self.current_mini_batches = {}
        self.current_grads        = {}
        self.total_raw            = 0
        self.total_useful         = 0

    def new_batch(self, t):
        data, targets = self.batch_mgr.get_next_batch()
        self.current_batch_id += 1
        minis = []
        for i in range(0, data.size(0), self.mini_batch_size):
            minis.append((data[i:i+self.mini_batch_size],
                          targets[i:i+self.mini_batch_size]))
        self.current_mini_batches = {i: minis[i] for i in range(len(minis))}
        self.current_grads = {}
        if self.verbose:
            print(f"[t={t}] Full batch {self.current_batch_id} ready ({len(minis)} mini-batches).")

    def register(self, grads, info, wid, t, ptime, is_byz=False):
        bid, midx = info
        if bid != self.current_batch_id: return
        if midx in self.current_grads: return
        self.total_raw    += ptime
        self.total_useful += ptime
        self.current_grads[midx] = grads

    def aggregate_and_update(self):
        if len(self.current_grads) != len(self.current_mini_batches):
            return False
        flat = {}
        for n in self.model.state_dict().keys():
            mats  = [self.current_grads[i][n].view(-1) for i in self.current_mini_batches]
            stack = torch.stack(mats)
            if   self.preagg == "NNM":       pa = byzfl.NNM(f=4)(stack)
            elif self.preagg == "Bucketing": pa = byzfl.Bucketing(s=1)(stack)
            elif self.preagg == "ARC":       pa = byzfl.ARC(f=4)(stack)
            else:                            pa = stack
            flat[n] = pa
        aggs = {}
        for n, pa in flat.items():
            if   self.aggregation_strategy == "trimmed_mean":   ag = byzfl.TrMean(f=4)(pa)
            elif self.aggregation_strategy == "multi_krum":     ag = byzfl.MultiKrum(f=4)(pa)
            elif self.aggregation_strategy == "geometric_median": ag = byzfl.GeometricMedian()(pa)
            elif self.aggregation_strategy == "average":        ag = byzfl.Average()(pa)
            elif self.aggregation_strategy == "meamed":         ag = byzfl.Meamed(f=4)(pa)
            else: raise ValueError(f"Unknown strategy {self.aggregation_strategy}")
            aggs[n] = ag.view(self.current_grads[0][n].shape)
        self.optimizer.zero_grad()
        for n,p in self.model.named_parameters():
            p.grad = aggs[n]
        self.optimizer.step()
        return True

    def evaluate(self, loader):
        self.model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for data, targets in loader:
                data, targets = data.to(self.device), targets.to(self.device)
                out = self.model(data)
                _, pred = torch.max(out, 1)
                total  += targets.size(0)
                correct += (pred == targets).sum().item()
        return 100 * correct / total

    def advance_decision(self, p1, p2, p3, sp):
        if p1 == p2 == p3: return "rollback"
        if p1>sp and p2>p1 and p3>sp: return "accept"
        if p1>sp and p2<=p1 and p3>p1: return "accept"
        if p1<=sp and p2>sp and p3>p2: return "accept"
        if p1<=sp and p2>p1 and p2>sp and p3<p2 and p3>sp: return "accept"
        if p1<=sp and p2<=p1 and p3>sp: return "accept"
        if p1 < sp and p2 < sp and p3 > sp: return "accept"
        return "rollback"

    def report(self):
        waste = self.total_raw - self.total_useful
        rate  = 100 * waste / self.total_raw if self.total_raw>0 else 0
        print(f"Raw={self.total_raw}, Useful={self.total_useful}, Waste={waste}, Rate={rate:.2f}%")

# ------------------------------------------------------------------------------
def create_workers(N, byz_ids, steps_per_cycle, server):
    ws = []
    for i in range(1, N+1):
        pt = random.randint(2,5)
        if i in byz_ids:
            av = generate_gaussian_byzantine_availability(steps_per_cycle, i, N, min_processing_time=pt)
            ws.append(ByzantineWorker(server.model, server.device, i, av, pt))
        else:
            av = generate_gaussian_availability(steps_per_cycle, i, N, min_processing_time=pt)
            ws.append(Worker(server.model, server.device, i, av, pt))
    return ws

# ------------------------------------------------------------------------------
class SafePointSimulator:
    def __init__(self, server, workers, test_loader, total_time_steps=1000):
        self.server           = server
        self.workers          = workers
        self.test_loader      = test_loader
        self.total_time_steps = total_time_steps
        self.global_time      = 0
        #self.plot_worker_availability()

    def plot_worker_availability(self):
        num_workers = len(self.workers)
        fig_height  = max(6, min(0.05 * num_workers, 20))
        fig, ax     = plt.subplots(figsize=(12, fig_height))
        for w in self.workers:
            c = "tab:red" if w.is_byzantine else "tab:blue"
            for (s, e) in w.availability:
                ax.broken_barh([(s, e - s)], (w.worker_id - 0.4, 0.8), facecolors=c)
        ax.set_xlabel("Temps")
        ax.set_ylabel("Worker ID")
        ax.set_title("Disponibilité des Workers (bleu=honnêtes, rouge=byzantins)")
        ax.set_xlim(0, self.total_time_steps)
        ax.set_ylim(0, max(w.worker_id for w in self.workers)+1)
        ax.grid(True, axis='y', linestyle='--', alpha=0.2)
        from matplotlib.patches import Patch
        ax.legend([Patch(facecolor='tab:blue'), Patch(facecolor='tab:red')],
                  ['Honnête','Byzantin'], loc='upper right')
        plt.tight_layout()
        plt.show()

        # cumul
        avail_h = [0]*self.total_time_steps
        avail_b = [0]*self.total_time_steps
        for w in self.workers:
            tgt = avail_b if w.is_byzantine else avail_h
            for s,e in w.availability:
                for t in range(s, min(e, self.total_time_steps)):
                    tgt[t] += 1
        plt.figure(figsize=(12,6))
        plt.plot(range(self.total_time_steps), avail_h, label="Honnêtes", linewidth=2)
        plt.plot(range(self.total_time_steps), avail_b, label="Byzantins", linewidth=2)
        plt.xlabel("Temps")
        plt.ylabel("Nombre disponibles")
        plt.title("Disponibilité cumulée des workers")
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()

    def run(self):
        safe_sd  = copy.deepcopy(self.server.model.state_dict())
        safe_opt = copy.deepcopy(self.server.optimizer.state_dict())
        sp_acc   = self.server.evaluate(self.test_loader)
        print(f"SafePoint initial — Acc: {sp_acc:.2f}%")

        current_sd, current_opt = safe_sd, safe_opt

        # première visualisation

        while self.global_time < self.total_time_steps:
            p_list = []
            incomplete = False

            for _ in range(3):
                if self.global_time >= self.total_time_steps:
                    incomplete = True
                    break

                # préparer batch
                self.server.model.load_state_dict(current_sd)
                self.server.optimizer.load_state_dict(current_opt)
                self.server.new_batch(self.global_time)

                data, targets = self.server.batch_mgr.get_next_batch()
                mini_batches = {}
                for i in range(0, data.size(0), self.server.mini_batch_size):
                    mini_batches[len(mini_batches)] = (data[i:i+self.server.mini_batch_size],
                                                       targets[i:i+self.server.mini_batch_size])
                M  = len(mini_batches)
                t0 = self.global_time

                def simulate_branch(sd, opt, pool, t0):
                    self.server.model.load_state_dict(sd)
                    self.server.optimizer.load_state_dict(opt)
                    self.server.current_mini_batches = mini_batches.copy()
                    self.server.current_grads        = {}
                    t, in_prog, used = t0, [], set()
                    while len(self.server.current_grads) < M and t < self.total_time_steps:
                        for w in list(in_prog):
                            if w.has_finished(t):
                                midx = w.current_info[1]
                                grads, _ = w.compute(*mini_batches[midx])
                                self.server.register(grads, (self.server.current_batch_id, midx),
                                                     w.worker_id, t, w.proc_time, w.is_byzantine)
                                used.add(w.worker_id)
                                in_prog.remove(w)
                        avail = [w for w in pool if w.is_available(t)
                                 and w not in in_prog and w.worker_id not in used]
                        for midx in sorted(mini_batches):
                            if midx in self.server.current_grads: continue
                            if not avail: break
                            w = avail.pop(0)
                            w.update_model(copy.deepcopy(sd))
                            d, tgt = mini_batches[midx]
                            w.start(t, (self.server.current_batch_id, midx), d, tgt)
                            in_prog.append(w)
                            used.add(w.worker_id)
                        t += 1
                    if len(self.server.current_grads) < M:
                        return used, None, None, None, None
                    self.server.aggregate_and_update()
                    acc = self.server.evaluate(self.test_loader)
                    return used, t, acc, \
                           copy.deepcopy(self.server.model.state_dict()), \
                           copy.deepcopy(self.server.optimizer.state_dict())

                # Branche A
                poolA = [copy.deepcopy(w) for w in self.workers]
                usedA, tA, accA, sdA, optA = simulate_branch(current_sd, current_opt, poolA, t0)
                if tA is None:
                    incomplete = True
                    break
                print(f"Branche A — Acc: {accA:.2f}% (took {tA-t0} steps)")

                # Branche B
                poolB = [copy.deepcopy(w) for w in self.workers if w.worker_id not in usedA]
                usedB, tB, accB, sdB, optB = simulate_branch(current_sd, current_opt, poolB, t0)
                if tB is None:
                    incomplete = True
                    break
                print(f"Branche B — Acc: {accB:.2f}% (took {tB-t0} steps)")

                # mettre à jour time comme max
                self.global_time = max(tA, tB)
                print(f"Temps ({self.global_time})")
                # choix du meilleur
                if accA >= accB:
                    current_sd, current_opt = sdA, optA
                    chosen_acc = accA
                else:
                    current_sd, current_opt = sdB, optB
                    chosen_acc = accB

                p_list.append(chosen_acc)

            if incomplete or len(p_list) < 3:
                print("Cycle incomplet → rollback et arrêt.")
                self.server.model.load_state_dict(safe_sd)
                self.server.optimizer.load_state_dict(safe_opt)
                break

            decision = self.server.advance_decision(p_list[0], p_list[1], p_list[2], sp_acc)
            print(f"\nadvance_decision({p_list}, sp={sp_acc:.2f}) → {decision}")
            if decision == "accept":
                safe_sd, safe_opt, sp_acc = current_sd, current_opt, p_list[-1]
                print("→ Accept : safe point mis à jour.")
            else:
                current_sd, current_opt = safe_sd, safe_opt
                print("→ Rollback : safe point restauré.")

        # fin
        self.server.model.load_state_dict(safe_sd)
        self.server.optimizer.load_state_dict(safe_opt)
        final_acc = self.server.evaluate(self.test_loader)
        print(f"\nAccuracy finale : {final_acc:.2f}%")
        return final_acc

# ------------------------------------------------------------------------------
def run_experiment(num_cycles=2, num_runs=1, steps_per_cycle=1000,
                   byzantine_ratio=0.0, aggregation_strategy='trimmed_mean', preagg=None):
    transform   = transforms.Compose([transforms.ToTensor(),
                                      transforms.Normalize((0.1307,), (0.3081,))])
    train_ds    = datasets.MNIST('./data', train=True, download=True, transform=transform)
    test_ds     = datasets.MNIST('./data', train=False, download=True, transform=transform)
    test_loader = DataLoader(test_ds, batch_size=64, shuffle=False)

    final_accs = []
    for run in range(num_runs):
        server = Server(train_ds, batch_size=640, mini_batch_size=64,
                        lr=0.005,
                        aggregation_strategy=aggregation_strategy,
                        preagg=preagg,
                        verbose=False)
        byz_ids = set(random.sample(range(1, 201), int(200 * byzantine_ratio)))
        workers = create_workers(200, byz_ids, steps_per_cycle, server)

        cycle_accs = []
        for cycle in range(num_cycles):
            print(f"\n===== Cycle {cycle+1}/{num_cycles} =====")
            sim = SafePointSimulator(server, workers, test_loader,
                                     total_time_steps=steps_per_cycle)
            acc = sim.run()
            server.report()
            cycle_accs.append(acc)
        final_accs.append(cycle_accs[-1])

    mean_acc = np.mean(final_accs)
    print(f"\nAccuracy moyenne sur tous les runs : {mean_acc:.2f}%")
    return mean_acc

# ------------------------------------------------------------------------------
if __name__ == "__main__":
    preagg_methods = [None]#, "NNM", "Bucketing", "ARC"]
    agg_methods    = ["average"]#, "trimmed_mean", "multi_krum", "geometric_median", "meamed"]
    byz_ratios     = [0.0, 0.1, 0.2, 0.3, 0.4]
    num_runs       = 1
    num_cycles     = 2
    steps_per_cycle= 1200

    for pre in preagg_methods:
        plt.figure(figsize=(10,7))
        for agg in agg_methods:
            results = []
            for ratio in byz_ratios:
                acc = run_experiment(num_cycles=num_cycles,
                                     num_runs=num_runs,
                                     steps_per_cycle=steps_per_cycle,
                                     byzantine_ratio=ratio,
                                     aggregation_strategy=agg,
                                     preagg=pre)
                results.append(acc)
            plt.plot([r*100 for r in byz_ratios], results,
                     marker='o', linewidth=2, markersize=8, label=agg)
        plt.title(f"Preagg={pre}")
        plt.xlabel("Taux byzantins (%)")
        plt.ylabel("Accuracy moyenne (%)")
        plt.legend(title="Agrégateur")
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
```
